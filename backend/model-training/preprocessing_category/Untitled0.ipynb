```python
# Drop rows with missing values in any of the three columns
df_cleaned = df.dropna(subset=["title", "category", "sub category"])
```

```python
# Define mappings for category and sub category
category_mapping = {
    "Politics & Governance": "PoliticsAndGovernance",
    "Science and Technology": "ScienceAndTechnology",
    "Culture & Entertainment": "CultureAndEntertainment",
    "Sports": "Sports",
    "Crime and Justice": "CrimeAndJustice"
}

sub_category_mapping = {
    "International Affairs": "InternationalPolitics",
    "Domestic Politics": "DomesticPolitics",
    "Tech & Innovation": "TechAndInnovation",
    "Research & Space": "ResearchAndSpace",
    "\xa0Screen & Stage": "ScreenAndStage",  # Handles non-breaking space
    "Music & Arts": "MusicAndArts",
    "Cricket": "Cricket",
    "Football": "Football",
    "Other Sports": "Other",
    "Crime Reports": "CrimeReport",
    "Courts & Investigations": "CourtsAndInvestigation"
}
```

```python
# Apply mappings
df_cleaned["category"] = df_cleaned["category"].map(category_mapping).fillna(df_cleaned["category"])
df_cleaned["sub category"] = df_cleaned["sub category"].map(sub_category_mapping).fillna(df_cleaned["sub category"])
```

```python
# Display unique values to confirm changes
df_cleaned["category"].unique(), df_cleaned["sub category"].unique()
```

```python
import re

# Function to remove non-Sinhala characters and numerical values
def clean_sinhala_text(text):
    # Sinhala Unicode range: U+0D80–U+0DFF
    # Retain Sinhala letters and basic punctuation/spaces
    return re.sub(r"[^\u0D80-\u0DFF\s]", "", text)
```

```python
# Apply the cleaning function to the 'title' column
df_cleaned["title"] = df_cleaned["title"].apply(clean_sinhala_text)
```
