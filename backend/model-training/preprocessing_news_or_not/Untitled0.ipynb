```python
# Drop rows with missing values in any of the three columns
df_cleaned = df.dropna(subset=["text", "label"])
```

```python
df_cleaned.info(), df_cleaned.head()
```

```python
df_cleaned['label'].unique()
```

```python
import re

# Function to remove non-Sinhala characters and numerical values
def clean_sinhala_text(text):
    # Sinhala Unicode range: U+0D80–U+0DFF
    # Retain Sinhala letters and basic punctuation/spaces
    return re.sub(r"[^\u0D80-\u0DFF\s]", "", text)
```

```python
# Apply the cleaning function to the 'title' column
df_cleaned["text"] = df_cleaned["text"].apply(clean_sinhala_text)
```

```python
cleaned_file_path="/content/drive/MyDrive/fnd-sample-1/classification_news_model/model-pipeline-news/news_sinhala_cleaned.csv"
df_cleaned.to_csv(cleaned_file_path, index=False, encoding='utf-8-sig')
```
